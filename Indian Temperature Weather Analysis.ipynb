### Dataset Prepration

Dataset Prepration Is Done By Using Capital List Of Indian States And From Raw data We Are Getting Temperature,City And Weather


```python
import requests
import json
temp = []
city = []
weather = []
k = 0
for i in caps:
    z=requests.get('http://api.openweathermap.org/data/2.5/weather?q='+i+'&appid=Your_API_Key')
    dic=json.loads(z.content.decode())
    if dic['cod'] == 200:
        k+=1
        print(k)
        print('City Name:',dic['name'])
        print('Temperature:',dic['main']['temp'])
        print('Weather:',dic['weather'][0]['main'])
        temp.append(dic['main']['temp'])
        city.append(dic['name'])
        weather.append(dic['weather'][0]['main'])
        
    else:
        print(i,'Spelling is Incorrect')
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.plot(temp,city,color='green', linestyle='dashed', linewidth = 3, 
         marker='o', markerfacecolor='blue', markersize=10)
plt.xlabel('Temperature In Kelvin(K)')
plt.ylabel('State Capitals Of India')
plt.title('Temperature Variation By State Capitals Of IN')
plt.show()
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.barh(city,temp)
plt.xlabel('Temperature In Kelvin(K)')
plt.ylabel('State Capitals Of India')
plt.title('Temperature Variation By State Capitals Of IN')
plt.show()
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
plt.scatter(temp,city,alpha=0.8,cmap='green',marker='*',edgecolors='green',linewidths=1.0,c='blue',s=100)
plt.xlabel('Temperature In Kelvin(K)')
plt.ylabel('State Capitals Of India')
plt.title('Temperature Variation By State Capitals Of IN')
plt.show()
```

```python
f,ax = plt.subplots(figsize = (9,15))
sns.barplot(x=data['Temp'][data['Weather']=='Clear'],y=data['City'],color='#3498DB',alpha = 0.5,label='Clear' )
sns.barplot(x=data['Temp'][data['Weather']=='Clouds'],y=data['City'],color='#FF4848',alpha = 0.7,label='Clouds')
sns.barplot(x=data['Temp'][data['Weather']=='Haze'],y=data['City'],color='#2ecc72',alpha = 0.6,label='Haze')
sns.barplot(x=data['Temp'][data['Weather']=='Mist'],y=data['City'],color='#FAD02E',alpha = 0.6,label='Mist')
sns.barplot(x=data['Temp'][data['Weather']=='Rain'],y=data['City'],color='#758AA2',alpha = 0.6,label='Rain')

ax.legend(loc='lower right',frameon = True)     # legendlarin gorunurlugu
ax.set(xlabel='Temperature', ylabel='States',title = "Temerature of State's According to Weather ")
```
